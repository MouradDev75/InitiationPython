```python
# supprimer le symbôle $ dans la colonne price_paid
def remplacer(value):
    return value.replace('$','')

df['price_paid'] = df['price_paid'].apply(remplacer)
df['price_paid']
```

```python
# modifier le type d'une colonne
df['price_paid'] = df['price_paid'].astype(float)
df['price_paid']
```
